# **Types Of Loops In Python **

```python
from os import umask
# 1. For Loop
#for variable in range(start, end):
  #repeated code here
for i in range(1, 6):
    print(i)

```

```python
# 2. While Loop
# while condition
i = 1
while i <= 5:
  print(i)
  i += 1
```

```python
# 3. Nested Loops
for i in range(1, 3):
    for j in range(1, 4):
        print(f"i={i}, j={j}")

```
